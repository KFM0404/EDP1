## Solución del problema de Dirichlet - Inciso (A)

La solución de la ecuación 
$$\frac{\partial^{2}u}{\partial x^{2}} + \frac{\partial^2 u}{\partial y^2}=0,\;\;\;0<x<1,\;\;\;0<y<1$$
sujeta a las condiciones de frontera:
$$u(0,y)=u(1,y)=0$$
$$u(x,0)=100, \;\;\; u(x,1)=200$$

La solución dada en términos de la serie de Fourier es:
$$u(x,y)=\sum_{n=1}^{\infty}\left[\left(\frac{200(1+(-1)^{n+1})}{n\pi}\right)cosh(n\pi y)+\left(\frac{200(1+(-1)^{n+1})}{n\pi}\right)\left(\frac{2-cosh(n\pi)}{senh(n\pi)}\right)senh(n\pi x)\right]sen(n\pi x)$$

```python
from pylab import *
```

```python
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)
```

```python
def u_x_y(x,y):
    N = 10
    f = zeros ((len(x),len(y)))
    for i in range(len(y)):
        for n in range(1,N):
          a = (200*(1+(-1)^(n+1)))/(n*pi)
          b = (200*(1+(-1)^(n+1))*(2-cosh(n*pi)))/((n*pi)*sinh(n*pi))
          f[:,i] += (a*cosh(n*pi*y[i])+b*sinh(n*pi*y[i]))*sin(n*pi*x)
    return f
```

Podemos ver esto en una gráfica en 3D.

```python
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$x$'); plt.ylabel('$y$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="viridis"); #Otros colores:"viridis", "inferno", "plasma"
```
