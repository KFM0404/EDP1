## Solución del problema de Dirichlet - Inciso (B)

La solución de la ecuación 
$$\frac{\partial^{2}u}{\partial x^{2}} + \frac{\partial^2 u}{\partial y^2}=0,\;\;\;0<x<1,\;\;\;0<y<1$$
sujeta a las condiciones de frontera:
$$u(0,y)=u(1,y)=0$$
$$u(x,0)=100, \;\;\; u(x,1)=200$$

La solución dada en términos de la serie de Fourier es:
$$u(x,y)=\sum_{n=1}^{\infty}\left[\frac{20(-1)^{n+1}}{n\pi}cosh(n\pi x)+ \left(\frac{2[(-1)^{n}-1]senh(n\pi)}{n^{2}\pi^{2}}+\frac{20(-1)^{n}}{n\pi}tanh(n\pi)\right)senh(n\pi x)\right]sen(n\pi y)$$

```python
from pylab import *
```

```python
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)
```

```python
def u_x_y(x,y):
    N = 10
    f = zeros ((len(x),len(y)))
    for i in range(len(x)):
        for n in range(1,N):
          a = (-20*(-1)**n)/(n*pi)
          b= (2*((-1)**n-1)+20*(n*pi)*sinh(n*pi)*(-1)**n)/(n**2*pi**2*cosh(n*pi))
          f[:,i] += (a*cosh(n*pi*x[i])+b*sinh(n*pi*x[i]))*sin(n*pi*y)
    return f
```

Podemos ver esto en una gráfica en 3D.

```python
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$x$'); plt.ylabel('$y$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="viridis"); #Otros colores:"viridis", "inferno", "plasma"
```
